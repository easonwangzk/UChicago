```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, make_scorer
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/easonwang/Desktop/leadership/W6/Verizon.csv")  # Update with your file path if needed

# Cleaning------------------------------------------------------
# Price doesn't equal 0
df = df[df['price'] != 0]
# Monthly Payment isn't greater than Price
df = df[df['monthly_payment'] < df['price']]
# Age is somewhat reasonable
df = df[~(df['age'] < 16)]
df['loan_to_value'] = (df['price'] - df['downpmt']) / df['price']
# Cleaning------------------------------------------------------

X = df.drop('default', axis=1)  # Features
y = df['default']  # Target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


# Custom loss function with binary crossentropy and additional term to guide optimization
def custom_loss(y_true, y_pred):
    bce = tf.keras.losses.BinaryCrossentropy()(y_true, y_pred)
    y_pred_binary = tf.sigmoid(y_pred)
    tp = tf.reduce_sum(y_true * y_pred_binary)
    fp = tf.reduce_sum((1 - y_true) * y_pred_binary)
    tn = tf.reduce_sum((1 - y_true) * (1 - y_pred_binary))
    fn = tf.reduce_sum(y_true * (1 - y_pred_binary))
    custom_term = 250 * (tn - fp) + 1000 * (tp - fn)
    return bce - 0.001 * custom_term

# Model 1: Improved Neural Network with Dropout and Batch Normalization
model_1 = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss=custom_loss, metrics=['accuracy'])
model_1.fit(X_train, y_train, epochs=100, batch_size=64, verbose=1, validation_split=0.2)
y_pred_1 = model_1.predict(X_test)
y_pred_binary_1 = np.round(y_pred_1)
conf_matrix_1 = confusion_matrix(y_test, y_pred_binary_1)

# Feature Importance for Model 1 using weights
weights_1 = model_1.get_weights()[0]  # Get weights of the first layer
feature_importance_1 = np.sum(np.abs(weights_1), axis=1)
importance_df_1 = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importance_1})
importance_df_1 = importance_df_1.sort_values(by='Importance', ascending=False)
plt.figure(figsize=(10, 6))
plt.barh(importance_df_1['Feature'], importance_df_1['Importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance for Model 1')
plt.gca().invert_yaxis()
plt.show()

# Model 2: Improved Deep Neural Network with Dropout and Batch Normalization
model_2 = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_2.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss=custom_loss, metrics=['accuracy'])
model_2.fit(X_train, y_train, epochs=100, batch_size=64, verbose=1, validation_split=0.2)
y_pred_2 = model_2.predict(X_test)
y_pred_binary_2 = np.round(y_pred_2)
conf_matrix_2 = confusion_matrix(y_test, y_pred_binary_2)

# Feature Importance for Model 2 using weights
weights_2 = model_2.get_weights()[0]  # Get weights of the first layer
feature_importance_2 = np.sum(np.abs(weights_2), axis=1)
importance_df_2 = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importance_2})
importance_df_2 = importance_df_2.sort_values(by='Importance', ascending=False)
plt.figure(figsize=(10, 6))
plt.barh(importance_df_2['Feature'], importance_df_2['Importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance for Model 2')
plt.gca().invert_yaxis()
plt.show()

# Model 3: XGBoost Model with Custom Scoring Function
# Custom scoring function for XGBoost to maximize (250 * (tp - fp) + 2500 * (tn - fn))
def custom_scorer(y_true, y_pred):
    y_pred_binary = (y_pred >= 0.5).astype(int)
    tp = np.sum((y_true == 1) & (y_pred_binary == 1))
    fp = np.sum((y_true == 0) & (y_pred_binary == 1))
    tn = np.sum((y_true == 0) & (y_pred_binary == 0))
    fn = np.sum((y_true == 1) & (y_pred_binary == 0))
    return 250 * (tn - fp) + 1000 * (tp - fn)

scorer = make_scorer(custom_scorer, greater_is_better=True)
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
grid_params = {
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5],
    'n_estimators': [100, 200]
}

grid_search = GridSearchCV(estimator=xgb_model, param_grid=grid_params, scoring=scorer, cv=3, verbose=1)
grid_search.fit(X_train, y_train)
y_pred_3 = grid_search.best_estimator_.predict(X_test)
conf_matrix_3 = confusion_matrix(y_test, y_pred_3)
print("Confusion Matrix for Model 3 (XGBoost):\n", conf_matrix_3)

# Feature Importance for Model 3
plt.figure(figsize=(10, 6))
xgb.plot_importance(grid_search.best_estimator_, importance_type='weight', show_values=False, title='Feature Importance for Model 3')

#contrast RF
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

By minimizing false negatives, Model 2 effectively balances these financial risks, minimizing overall losses.
